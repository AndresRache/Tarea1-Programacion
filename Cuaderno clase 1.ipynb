```python
class DataExhaustedError(Exception):
    pass

class IrisDataIterator:
    def __init__(self, file_path):
        self.data = pd.read_csv(file_path)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index + 1 < len(self.data):
            row = self.data.iloc[self.index:self.index + 2]
            self.index += 2
            return row
        else:
            raise DataExhaustedError("No hay más datos disponibles.")

# Uso de la clase iterable
if __name__ == "__main__":
    iris_iterator = IrisDataIterator("iris.csv")

    try:
        for batch in iris_iterator:
            print(batch)
    except DataExhaustedError as e:
        print(e)
```
